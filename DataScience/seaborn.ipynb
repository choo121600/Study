# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.info()
```

## relplot()

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind='line')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='time')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, col='time')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='day')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, col='day')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='size')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='size', palette='coolwarm')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, style='size')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, size='size')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, size='size', sizes=(15, 200))
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='time', style = 'sex', size='size')
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue='time', style = 'sex', size='size')
```

## lmplot()

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

## regplot()

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

## lmplot vs regplot

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker", col="time")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker", col="time", row="sex")
```

## jointplot()

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, hue='time')
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, hue='time', kind='kde')
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
```

## catplot()

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="box")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="box", hue="smoker")
```

```python
sns.catplot(x="total_bill", y="day", data=tips, kind="violin", hue="sex")
```

```python
sns.catplot(x="total_bill", y="day", data=tips, kind="violin", hue="sex", split=True)
```

## pairplot()

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```
